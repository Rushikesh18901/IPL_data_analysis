```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/matches.csv")

# Total matches played
print(" Total IPL Matches Played:", df.shape[0])   # returns the total no. of rows in the dataset mathces in the dataset

# Top 5 teams with most wins
team_wins = df['winner'].value_counts().head(5)   # how many times each team has own  , top 5 teams
print("\n Top 5 Teams by Wins:\n", team_wins)

```

```python
# Average wins of top 5 teams
avg_wins = np.mean(team_wins.values)                              # calculating the avg no of wins
print(f"\n Average wins among Top 5 Teams: {avg_wins:.2f}")

#Team with highest wins using argmax
index_of_max = np.argmax(team_wins.values)   # gives the index of the max value
team_with_max_wins = team_wins.index[index_of_max]
print(f" Team with Highest Wins : {team_with_max_wins}")

```

```python
# Plot: Most Matches Won
team_wins.plot(kind='bar', color='skyblue')   # visualizing the top 5 winning teams
plt.title("Top 5 Teams by Match Wins")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()

```

```python
# Top 5 Player of the Match winners
top_players = df['player_of_match'].value_counts().head(5) # calculating the individual performance of the player
print("\n Top 5 Players of the Match:\n", top_players)


```

```python
# Plot: Player of the Match
sns.barplot(x=top_players.values, y=top_players.index, palette='viridis')
plt.title("Top 5 Player of the Match Winners")
plt.xlabel("Awards")
plt.ylabel("Players")
plt.show()

```

```python
# Plot: Player of the Match - Pie Chart
plt.pie(top_players.values, labels=top_players.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(top_players)))
plt.title("Top 5 Player of the Match Winners")
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.show()
```

```python
# Final Summary
print("\n Summary:")
print(f"- Total Matches: {df.shape[0]}")
print(f"- Most Successful Team: {df['winner'].value_counts().idxmax()}")
print(f"- Top Player of the Match: {df['player_of_match'].value_counts().idxmax()}")
```
